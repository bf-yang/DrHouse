```python
# split our collected dataset into training and test sets
import pandas as pd
import os, random
import torch
import numpy as np

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True

random_seed = 1999
setup_seed(random_seed)

# split the dataset and save as training and test sets.
df = pd.read_excel("./adaptive_retrieval_dataset.xlsx", sheet_name='dataset')
df = df.sample(frac=1.0, random_state=42).reset_index()
train_df = df.sample(frac=0.7, random_state=42)
test_df = df[~df.index.isin(train_df.index)]

filename_train = 'train.xlsx'
filename_test = 'test.xlsx'
train_df.to_excel(filename_train, sheet_name='dataset', index=False)
test_df.to_excel(filename_test, sheet_name='dataset', index=False)
```
